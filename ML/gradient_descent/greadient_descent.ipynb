<!-- gradient descent an optimisation algo. that finds the best fit line for the dataset -->

```python
# importing libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# using numpy arrays over python's list is faster
# input feature
x=np.array([1,2,3,4,5])
# output feature
y=np.array([5,7,9,11,13])
```

```python
# intial params can be of any values 0, +, - 
m=0 # intercept
b=0 # slope
learning_rate=0.01
epochs=1000 # iteration
n=len(x)
```

```python
for i in range(epochs):
    y_pred=m*x+b
    cost=(1/n)*sum([val**2 for val in (y-y_pred)]) # cost decreases as we reach near the minima
    dm=(-2/n)*np.sum(x*(y-y_pred))
    db=(-2/n)*np.sum(y-y_pred)
    m=m-learning_rate*dm
    b=b-learning_rate*db
    
print("Slope (m):", m)
print("Intercept (b):", b)
print("Cost (c):", cost)
```

```python
plt.scatter(x, y)
plt.plot(x, m * x+ b)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression using Gradient Descent")
plt.show()
```
