*** Vectorizer and Pipeline ***

üëâ It converts text ‚Üí numbers by counting words.

Example

Sentence: "free offer free"

Vocabulary created:

    free  ‚Üí index 0

    offer ‚Üí index 1

Vector:

[2, 1]


‚úî free appears 2 times

‚úî offer appears 1 time

üîç Example in Code

from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X = v.fit_transform(["free offer free"])

‚öôÔ∏è Why We Use fit() and transform()
Method	Meaning

| Method            | Meaning                 |
| ----------------- | ----------------------- |
| `fit()`           | Learn vocabulary        |
| `transform()`     | Convert text to numbers |
| `fit_transform()` | Both together           |

Rule:

Fit only on training data

üî¥ Problem Without Pipeline

X_train_count = v.fit_transform(X_train)

X_test_count = v.transform(X_test)

model.fit(X_train_count, y_train)

model.predict(v.transform(new_email))


‚ùå Too many steps

‚ùå Easy to make mistakes

‚ùå Re-fitting accidentally breaks model

‚úÖ Solution: sklearn Pipeline (BEST PRACTICE)

What is Pipeline?

Pipeline automates steps and ensures the same transformations are applied every time.

üß† Pipeline Flow

    Text ‚Üí CountVectorizer ‚Üí MultinomialNB ‚Üí Prediction

‚úÖ Clean Pipeline Code (Spam Classifier Example)

from sklearn.pipeline import Pipeline

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.naive_bayes import MultinomialNB

model = Pipeline([

        ('vectorizer', CountVectorizer()),

        ('nb', MultinomialNB())

])

‚úÖ Training (ONE LINE)

model.fit(X_train, y_train)


‚úî Vectorizer is fitted automatically

‚úî Model is trained automatically

‚úÖ Prediction (ONE LINE)

emails = [

    'Hey mohan, can we get together to watch football game tomorrow?',

    'Upto 20% discount on parking, exclusive offer just for you'

]

model.predict(emails)


‚úî No manual transform

‚úî No feature mismatch

‚úî No errors

üéØ Final thought

‚ÄúPipeline combines preprocessing and model into a single object, ensuring consistent transformations and preventing errors caused by manual transformations.‚Äù


Pipeline ensures:

Same vectorizer

Same vocabulary

Same feature size

‚úî Error eliminated forever

‚ö†Ô∏è Common Beginner Mistakes

‚ùå Using fit_transform() on test data

‚ùå Predicting without transforming

Pipeline solves all ‚ùå‚ùå‚ùå

```python
import pandas as pd
```

```python
df=pd.read_csv("spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.Message,df.spam,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train)
x_train_count.toarray()
```

![vectorizer](countVector.png)

here in the 4 documents these are unique words found and the matrix is formed based on the occurences of those words in the documents

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train) # here not the x_train bcoz it is txt x_train_count is vectorised ready for the model 
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails) # not fit_transform
model.predict(emails_count)
```

The SAME vectorizer used for training must be used for prediction

If vocabulary changes ‚Üí model breaks ‚ùå


‚ÄúIn text classification, the vectorizer must be fitted only on training data. For prediction, we use transform() with the same vectorizer to maintain feature consistency.‚Äù

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
# model is pipelined no transformation needed before using in model pipeline itself converts text into vector and gives to model
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
