* sklearn.datasets has some predefined datasets
* dir(obj) - shows methods, attributes of the object
* matshow - Takes a 2D array (list of or NumPy array) and provides the quick visualisation
* seaborn - another visualisation lib
* confusion matrix

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
dir(digits)
```

```python
# len(digits.data)
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# same first param is multidimentional array
# model.predict([digits.data[67]])
predicted=model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
model.predict(digits.data[0:5])
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(predicted,y_test)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

* confusion matrix - evaluate the performance of a classification model by actual +ves, -ves and predicted +ves and -ves, tells exactly where the model fails.It tells
What the model predicted

What the actual answer was

Where the model is correct

Where the model is making mistakes

Think of it as a report card for a classification model.
